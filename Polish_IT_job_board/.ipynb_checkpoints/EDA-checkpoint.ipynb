#### `Published_at` conversion

```python
#converting Published_at as datetime type
df['Published_at_datetime'] = df['Published_at'].astype('datetime64[s]')
```

```python
#checking whether we didn't lose any value due to conversion
df[df['Published_at_datetime'].isnull()].shape, df[df['Published_at'].isnull()].shape
```

##### Extracting dateparts

```python
df['Published_at_date'] = df.Published_at_datetime.dt.date
df['Published_at_time'] = df.Published_at_datetime.dt.time
df['Published_at_hour'] = df.Published_at_datetime.dt.hour
df['Published_at_day'] = df.Published_at_datetime.dt.day
df['Published_at_dayofyear'] = df.Published_at_datetime.dt.dayofyear
df['Published_at_weekday'] = df.Published_at_datetime.dt.weekday
df['Published_at_month'] = df.Published_at_datetime.dt.month
df['Published_at_year'] = df.Published_at_datetime.dt.year
```

#### NaN analysis

```python
#df with rows that contain a NaN
df_filtered_Nan = df[~df.notnull().all(1)]
```

```python
df_filtered_Nan.info()
```

##### `Open_to_hire_Ukrainians`

```python
df_filtered_Nan[df_filtered_Nan['Open_to_hire_Ukrainians'].isnull()].sort_values('Published_at').head(7)
```

```python
(min(df_filtered_Nan[df_filtered_Nan['Open_to_hire_Ukrainians'].isnull()].Published_at),
max(df_filtered_Nan[df_filtered_Nan['Open_to_hire_Ukrainians'].isnull()].Published_at))
```

```python
#checking min published date for non-NaN rows
min(df['Published_at'].loc[pd.notnull(df.Open_to_hire_Ukrainians)])
```

```python
df_filtered_Nan.groupby(['Open_to_hire_Ukrainians', 'Published_at_date'], dropna=False)['Open_to_hire_Ukrainians', 'Published_at_date'].count()
```

Conclusion: Russian invasion of Ukraine started on 24 February 2022. For a while there was no "Open to hire Ukrainians" filter. It seems like this field became mandatory after 2022-04-04.

#### `Company_size_from`, `Company_size_to`          

```python
df_filtered_Nan[df_filtered_Nan['Company_size_from'].isnull()].head(7)
```

```python
df_filtered_Nan.groupby(['Company_size_from'], dropna=False)['Company_size_from'].count()
```

Conclusion: We have some odd string values. There are also decimal and integer values.

```python
df_filtered_Nan.query('Company_size_from == "May"').head()
```

Conclusion: String values of `Company_size_from` might be a conversion error - on above examples it looks like there should be number 5 in `Company_size_from` column.

Let's check possible combinations of `Company_size_from` and `Company_size_to` columns.

```python
df_filtered_Nan.groupby(['Company_size_from','Company_size_to'], dropna=False)[['Company_size_from','Company_size_to']].count()
```

Conclusion: There's a lot of 0 values in `Company_size_to`. Let's check these records first.

```python
df_filtered_Nan.query('Company_size_to == "0"').head()
```

Conclusion: It's possible to put information only with minimum company size (e.g. 40+). I assume rows with positive `Company_size_from` and zero `Company_size_to` are these kinds of offers.

Let's check combinations with names of months in the column.

```python
invalid_Company_size_values_months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
df_filtered_Nan[df_filtered_Nan['Company_size_from'].isin(invalid_Company_size_values_months)].groupby(['Company_size_from','Company_size_to'],
dropna=False)[['Company_size_from']].count()
```

```python
df_filtered_Nan[df_filtered_Nan['Company_size_to'].isin(invalid_Company_size_values_months)].groupby(['Company_size_from','Company_size_to'],
dropna=False)[['Company_size_to']].count()
```

Conclusion: These are conversion errors. We will later replace these values with number on the main DataFrame.
[Go to: Replacing months values in Company_size_from and Company_size_to](#replacing-months-values-in-company-size-from-and-company-size-to)

Let' check out the rest of values.

```python
df_filtered_Nan[(df_filtered_Nan[['Company_size_from', 'Company_size_to']].isin(invalid_Company_size_values_months)) 
& (df_filtered_Nan['Company_size_to'] != "0")].groupby(['Company_size_from','Company_size_to'], 
dropna=False)[['Company_size_from','Company_size_to']].count()
```

Conclusion: There are only NaNs left.

#### Replacing months values in `Company_size_from` and `Company_size_to`

```python
df['Company_size_from'] = df['Company_size_from'].replace(invalid_Company_size_values_months, list(range(1,13)))
df['Company_size_to'] = df['Company_size_to'].replace(invalid_Company_size_values_months, list(range(1,13)))
```
